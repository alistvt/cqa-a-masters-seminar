```python
from torch.utils.data import Dataset, DataLoader

batch_size = 2

class CoQADataset(Dataset):
    def __init__(self, data):
        self.data = data

    def __getitem__(self, idx):
        return self.data[idx]
        # return {
        #     'story': self.data['questions'][idx],
        #     'questions': self.data['questions'][idx],
        #     'answers': self.data['answers'][idx],
        # }

    def __len__(self):
        return len(self.data)

    @staticmethod
    def collate_fn(batch):
        return batch


dataset_train = CoQADataset(coqa['train'])
dataset_val = CoQADataset(coqa['validation'])

loader_train = DataLoader(dataset_train, batch_size=batch_size, shuffle=True) # collate_fn=CoQADataset.collate_fn
loader_val = DataLoader(dataset_val, batch_size=batch_size, shuffle=True) # collate_fn=CoQADataset.collate_fn
```
